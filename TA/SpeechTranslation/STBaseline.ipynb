```python
data = raw_datasets["train"][:1000]
```

```python
hypotheses = []
for sample in data["file"]:
    hypotheses.append((model.transcribe(sample, language="English"))["text"])
```

```python
data["hypothesis"] = hypotheses
data["hypothesis"][:5]
```

```python
normalizer = BasicTextNormalizer()
data["hypothesis_clean"] = [normalizer(text) for text in data["hypothesis"]]
data["sentence_clean"] = [normalizer(text) for text in data["sentence"]]
data["translation_clean"] = [normalizer(text) for text in data["translation"]]
```

```python
from evaluate import load

metric = load("sacrebleu")
```

```python
result = metric.compute(
    predictions=data["hypothesis_clean"], references=data["translation_clean"]
)
print(f'BLEU score: {result["score"]:.1f}')
```

```python
from evaluate import load

comet_metric = load("comet")
```

```python
comet_score = comet_metric.compute(
    predictions=data["hypothesis_clean"],
    references=data["translation_clean"],
    sources=data["sentence_clean"],
)
```

```python
print(f"COMET: {comet_score['mean_score'] * 100:.2f} %")
```
